```python
import httpx
import pandas as pd
from pathlib import Path
```

```python
URL = "https://api.openf1.org/v1/"
BRONZE = Path.cwd().parent / "data" / "bronze"
```

```python
endpoints = [
    "drivers",
    "meetings",
    "sessions",
    "session_result",
    "laps",
    "pit",
    "stints",
    "weather",
    "race_control",
    "championship_drivers",
]
```

```python
def extract(endpoint):
    data = httpx.get(URL + endpoint, timeout=None).json()
    if not isinstance(data, list):
        print(endpoint, data)
        return None
    df = pd.DataFrame(data)
    df.to_parquet(BRONZE / f"bd_{endpoint}.parquet", index=False)
    return df
```

```python
tablas = {endpoint: extract(endpoint) for endpoint in endpoints}
```
