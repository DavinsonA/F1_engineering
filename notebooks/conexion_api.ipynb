```python
import httpx
import pandas as pd
from pathlib import Path
import shutil
import os
```

```python
ROOT = Path.cwd().parent

BRONZE = ROOT / "data" / "bronze"
```

```python
URL = "https://api.openf1.org/v1/"

driver = httpx.get(URL + "drivers")
session_result = httpx.get(URL + "session_result")
meetings = httpx.get(URL + "meetings")
sessions = httpx.get(URL + "sessions")
weather = httpx.get(URL + "weather")
stints = httpx.get(URL + "stints")
laps = httpx.get(URL + "laps", timeout=None)
pits = httpx.get(URL + "pit")
race_control = httpx.get(URL +"race_control")
champ_driver = httpx.get(URL + "championship_drivers")
```

```python
bd_drivers = pd.DataFrame(driver.json())
bd_sessions = pd.DataFrame(sessions.json())
bd_meetings = pd.DataFrame(meetings.json())
bd_weather = pd.DataFrame(weather.json())
bd_stints = pd.DataFrame(stints.json())
bd_laps = pd.DataFrame(laps.json())
bd_pits = pd.DataFrame(pits.json())
bd_race_control = pd.DataFrame(race_control.json())
bd_champ_driver = pd.DataFrame(champ_driver.json())
```

```python
bd_drivers.to_parquet(BRONZE / "bd_pilotos.parquet", index=False)
bd_sessions.to_parquet(BRONZE / "bd_sessions.parquet", index=False)
bd_meetings.to_parquet(BRONZE / "bd_metings.parquet", index=False)
bd_weather.to_parquet(BRONZE / "bd_weather.parquet", index=False)
bd_stints.to_parquet(BRONZE / "bd_stints.parquet", index=False)
bd_laps.to_parquet(BRONZE / "bd_laps.parquet", index=False)
bd_pits.to_parquet(BRONZE / "bd_pits.parquet", index=False) 
bd_race_control.to_parquet(BRONZE / "bd_race_control.parquet", index=False)
bd_champ_driver.to_parquet(BRONZE / "bd_champ_driver.parquet",index=False)
```

```python
def connection_api(endpoint):
    #conexion a la API
    endpoint_api = httpx.get(URL + endpoint)
    bd_endpoint = pd.DataFrame(endpoint_api.json())
    bd_endpoint.to_parquet(BRONZE / f"bd_{endpoint}.parquet", index=False)
    return bd_endpoint.head()
```
